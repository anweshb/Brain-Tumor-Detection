```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
import seaborn as sns
import random
```

```python
train_dir = 'brain_tumor_final//train//'
```

```python
normal_imgs = os.listdir(train_dir+ '0//')
tumor_imgs = os.listdir(train_dir + '1//')
```

```python
rand_norm = np.random.choice(normal_imgs, 3, replace = False)
rand_tumor = np.random.choice(tumor_imgs, 3, replace = False)
```

```python
rand_norm.shape
```

```python
rand_norm
```

```python
fig = plt.figure(figsize = (13,10))
for i in range(6):
    if i < 3:
        fp = f'{train_dir}/0/{rand_norm[i]}'
        label = 'NORMAL'
    else:
        fp = f'{train_dir}/1/{rand_tumor[i-3]}'
        label = 'TUMOR'
    ax = fig.add_subplot(2, 3, i+1)
    
    fn = image.load_img(fp, target_size = (200,200), color_mode='grayscale')
    plt.imshow(fn, cmap='gray',aspect='auto')
    plt.title(label)
    plt.axis('off')
plt.show()
```

```python
dim1 = []
dim2 = []
for sub_folders in os.listdir(train_dir):
    for imgs in os.listdir(train_dir + sub_folders):
        img = plt.imread(train_dir + sub_folders +'//' + imgs)
        dim1.append(img.shape[0])
        dim2.append(img.shape[1])
```

```python
dim1 = round(np.mean(dim1))
dim2 = round(np.mean(dim2))
```

```python
print("Average dimensions of image: ({},{})".format(dim1,dim2))
```
